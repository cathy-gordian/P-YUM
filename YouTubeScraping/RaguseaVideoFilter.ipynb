On his channel, Ragusea says he posts recipe videos on Thursdays, so we will kill all videos not posted on Thursday

```python
import pandas as pd
import datetime
```

```python
rag_df = pd.read_csv("AdamRaguseaVideos.csv")
```

```python
rag_df["weekday"] = pd.to_datetime(rag_df['date published']).dt.dayofweek
```

```python
rag_df.head()
```

```python
rag_df = rag_df[rag_df['weekday'] == 3]
```

```python
rag_df.head()
```

```python
rag_df.drop(columns=rag_df.columns[0], 
        axis=1, 
        inplace=True)
```

```python
rag_df.head()
```

```python
rag_df.to_csv("AdamRaguseaVideos.csv", index = False)
```
